# Bar Chart

**‚úÖ What is a Bar Chart?**

A bar chart displays categorical data with rectangular bars.

The length/height of the bar represents the value or frequency of that category.

**Example: Sales of Different Products**

üéØ Objective:

Visualize the sales of different products using a bar chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data: Product names and their sales
products = ['Laptop', 'Tablet', 'Smartphone', 'Headphones', 'Smartwatch']
sales = [150, 100, 300, 120, 80]
```

```python
# Create a bar chart
plt.bar(products, sales, color='skyblue', edgecolor='black')

# Add title and axis labels
plt.title('Sales of Electronic Products (in Units)')
plt.xlabel('Product')
plt.ylabel('Units Sold')

# Add grid and display the plot
plt.grid(axis='y')
plt.show()
```

**Explanation of the Code**
- 1Ô∏è‚É£ products: List of product categories (x-axis).
- 2Ô∏è‚É£ sales: Corresponding sales data (y-axis).
- 3Ô∏è‚É£ plt.bar(...):

    - Creates vertical bars for each product.

    - color='skyblue': Sets bar color.

    - edgecolor='black': Adds a black border to the bars.

- 4Ô∏è‚É£ plt.title(), plt.xlabel(), plt.ylabel(): Add labels and title.
- 5Ô∏è‚É£ plt.grid(axis='y'): Horizontal grid lines for better readability.
- 6Ô∏è‚É£ plt.show(): Displays the chart.

**üìä Output Description**

You‚Äôll see a bar chart where:

Each bar represents a product.

The height shows how many units were sold.

**Example: Students Enrolled in Courses**

üéØ Objective:

Visualize student enrollment in different courses using a bar chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data: Courses and student enrollments
courses = ['Python', 'Data Science', 'Web Dev', 'AI/ML', 'Cyber Security']
students = [120, 95, 75, 110, 60]
```

```python
# Create the bar chart
plt.bar(courses, students, color='green', edgecolor='black')

# Add title and labels
plt.title('Student Enrollment in Different Courses')
plt.xlabel('Courses')
plt.ylabel('Number of Students')

# Add grid and show the chart
plt.grid(axis='y')
plt.show()
```

**Explanation**
- 1Ô∏è‚É£ courses: List of course names (x-axis).
- 2Ô∏è‚É£ students: Number of students enrolled in each course (y-axis).
- 3Ô∏è‚É£ plt.bar(...): Draws vertical bars for each course.
- 4Ô∏è‚É£ plt.title(), plt.xlabel(), plt.ylabel(): Provide context to the chart.
- 5Ô∏è‚É£ plt.grid(axis='y'): Adds horizontal grid lines.
- 6Ô∏è‚É£ plt.show(): Renders the final chart.

**üìä Output**

A bar chart where each bar shows how many students are enrolled in each course ‚Äî making it easy to compare course popularity.
