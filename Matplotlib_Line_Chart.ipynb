# Line Charts

**âœ… What is a Line Chart?**

A line chart shows trends over time or ordered data points by connecting data points with a continuous line.

Itâ€™s ideal for time-series data, growth patterns, and comparisons.

**âœ… Example: Monthly Sales Trend**

ðŸŽ¯ Objective:

Visualize monthly sales (in â‚¹1000s) over a year using a line chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data: Months vs Sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [25, 28, 30, 35, 40, 38, 45, 50, 48, 42, 38, 33]
```

```python
# Create the line chart
plt.plot(months, sales, color='blue', marker='o', linestyle='-', linewidth=2)
```

```python
# Add title and axis labels
plt.title('Monthly Sales Trend (in â‚¹1000s)')
plt.xlabel('Month')
plt.ylabel('Sales (â‚¹1000s)')

# Add grid and display the plot
plt.grid(True)
plt.show()
```

**Explanation of the Code**
- 1ï¸âƒ£ months: List of month names (x-axis).
- 2ï¸âƒ£ sales: Corresponding sales data (y-axis).
- 3ï¸âƒ£ plt.plot(...):

    - color='blue': Line color

    - marker='o': Circle markers at each data point

    - linestyle='-': Solid line

    - linewidth=2: Line thickness

- 4ï¸âƒ£ plt.title(), plt.xlabel(), plt.ylabel(): Add a title and axis labels.
- 5ï¸âƒ£ plt.grid(True): Adds grid lines for better readability.
- 6ï¸âƒ£ plt.show(): Displays the chart.

**ðŸ“Š Output Description**

The line chart will show how sales increase, peak, and drop across the months.

Useful for identifying trends, seasonality, or business performance over time.

**Example: Sales Comparison of Product A vs Product B**

ðŸŽ¯ Objective:

Visualize and compare monthly sales trends of two products using a line chart.

```python
import matplotlib.pyplot as plt
```

```python
# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [30, 35, 40, 42, 45, 50]
sales_B = [25, 28, 35, 37, 40, 48]
```

```python
# Create line chart for Product A
plt.plot(months, sales_A, label='Product A', color='blue', marker='o', linestyle='-')

# Create line chart for Product B
plt.plot(months, sales_B, label='Product B', color='green', marker='s', linestyle='--')

# Add title and labels
plt.title('Monthly Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales (â‚¹1000s)')

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

**Explanation of the Code**
- 1ï¸âƒ£ months: X-axis labels for months.
- 2ï¸âƒ£ sales_A and sales_B: Y-axis data for Product A and

B sales.

- 3ï¸âƒ£ plt.plot(...):


Product A:

- color='blue', marker='o', linestyle='-'

Product B:

- color='green', marker='s', linestyle='--'

- 4ï¸âƒ£ plt.legend(): Adds a legend to differentiate the two products.
- 5ï¸âƒ£ plt.grid(True): Adds grid lines for better readability.
- 6ï¸âƒ£ plt.show(): Displays the chart.

**ðŸ“Š Output Description**

You will see two lines:

Product A in blue with circle markers

Product B in green with square markers

The chart allows you to compare growth trends side-by-side.
